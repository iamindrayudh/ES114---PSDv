Q1. List Operations and Basic Statistics:

Create a list of the first 20 natural numbers.

Compute the sum of all even numbers in the list.

Compute the product of all odd numbers in the list.

Calculate the mean, median, and standard deviation of the list.


```python
# Step 1: Create a list of the first 20 natural numbers
natural_numbers = list(range(1, 21))

# Step 2: Compute the sum of all even numbers in the list
even_sum = sum(num for num in natural_numbers if num % 2 == 0)

# Step 3: Compute the product of all odd numbers in the list
odd_product = 1
for num in natural_numbers:
    if num % 2 != 0:
        odd_product *= num

# Step 4: Calculate the mean
mean = sum(natural_numbers) / len(natural_numbers)

# Step 5: Calculate the median
sorted_numbers = sorted(natural_numbers)
mid = len(sorted_numbers) // 2
if len(sorted_numbers) % 2 == 0:
    median = (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
else:
    median = sorted_numbers[mid]

# Step 6: Calculate the standard deviation
variance = sum((num - mean) ** 2 for num in natural_numbers) / len(natural_numbers)
std_dev = variance ** 0.5

# Displaying the results
print("List of natural numbers:", natural_numbers)
print("Sum of even numbers:", even_sum)
print("Product of odd numbers:", odd_product)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

Q2. Mutable vs. Immutable Data Types:
Explain the difference between mutable and immutable data types in Python. Provide examples of each.

Mutable data types: 
These can be modified after their creation. You can change their content (add, update, or delete elements) without changing their memory reference.


```python
my_list = [1, 2, 3]
my_list.append(4)  # Adds 4 to the list
print(my_list)  # Output: [1, 2, 3, 4]
```

Immutable data types: 
These cannot be modified after their creation. Any attempt to modify them creates a new object in memory.

```python
my_tuple = (1, 2, 3)
# my_tuple[0] = 0  # This will raise a TypeError
new_tuple = my_tuple + (4, 5)
print(new_tuple)  # Output: (1, 2, 3, 4, 5)
```

Q3. Fibonacci Sequence:
Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
Plot the Fibonacci sequence using a line graph.


```python
import matplotlib.pyplot as plt

# Function to generate the first n Fibonacci numbers
def generate_fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence

# Number of Fibonacci numbers to generate
n = 20
fibonacci_numbers = generate_fibonacci(n)

# Plotting the Fibonacci sequence
plt.figure(figsize=(10, 6))
plt.plot(range(1, n + 1), fibonacci_numbers, marker='o', linestyle='-', color='blue', label='Fibonacci Sequence')
plt.title('Fibonacci Sequence', fontsize=16)
plt.xlabel('Index (n)', fontsize=12)
plt.ylabel('Fibonacci Number', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```
